# LISTAS EM PYTHON


```python
#para criarmos uma lista usaremos o simbolo [] colchetes
#essa lista precisa ser armazenada em uma VARIAVEl
#os itens de uma lista são separados pela virgula

#lista_compras é a variavel
#a nossa lista de compras é composta por 3 tens: banana, laranja e maça
lista_compras = ['banana','laranja','maçã']
print(lista_compras)
```

ATRIBUINDO 3 ITENS A VARIAVEL
LISTA_COMPRAS

```python
#CADA ITEM DA LISTA POSSUI UMA POSICAO. POR PADRAO, AS POSICOES NAO INICIAM DO 1(UM) E SIM DO 0 (ZERO)

#para acessarmos um item da lista vamos utilizar a estrutura: nomeda lista[posição]
# buscando a posição [1]
print(lista_compras[1])
```

```python
# buscando a posicao [0]
print(lista_compras[0])
```

```python
# se tentarmos  a posicao [3] teremos um erro. isso ocorre pois não existe uma posição 3. apenas as posições [0], [1], [2]
print(lista_compras[3])
```

```python
#o uso de numeros negativos permite acessar a lista na ordem inversa
print(lista_compras[-1])
```

ADICIONAR E DELETAR ITENS DE UMA LISTA

```python
#PARA EXCLUIRMOS UM ITEM DA LISTA PODEMOS ISAR ALGUMAS OPCOES
# - del
# - .pop()
# - .remove()

# .append() -> adiciona AO FIM da lista_compras o item 'carro'
lista_compras.append('carro')
print(lista_compras)
```

```python
# del -> remove um item da lista baseado na posição indicada
del lista_compras[3]
print(lista_compras)
```

```python
# .insert -> insere um item na lista na posição indicada
lista_compras.insert(1, 'carro')
print(lista_compras)
```

```python
lista_compras.append('carro')
```

```python
# .remove() -> remove um item baseado no seu valor e não na sua posição
#ATENÇÃO! retira apenas a primeira ocorrencia e não todas
lista_compras.remove('carro')
```

```python
print(lista_compras)
```

*append:* adicionar um item ao final da nossa lista.
*insert: *podemos escolher a posição exata em que vamos inserir esse item.
*remove:* nós temos a remoção do item de acordo com a sua escrita/valor e só a primeira ocorrência será removida e não todos!

## CRIANDO LISTA DO ZERO

```python
#É muito comum que as listas se iniciem sem nenhum valor, como se fosse um papel em branco que aos poucos voce vai adicionando informações
# para criarmos vamos usar novamente o colochete, mas em nenhum item

#lista vazia
lista_sonhos = []
# .pop() -> Remove da lista_compras o ultimo item, mas nao o exclui, nesse caso o valor carro será armazenado na variavel item
item = lista_compras.pop(-1)
print(item)
#apos retirar da lista_compras, adicionamos 'carro' na nossa lista_sonhos
lista_sonhos.append(item)
```

```python
print(lista_sonhos)
```

# CRIANDO UMA LISTA DE TAREFAS (TO D0)

```python
# criando a lista tarefas
tarefas = []
# usando o input() vamos coletar do usuario qual a tarefa a ser adicionada
atividade = input('INSIRA UMA ATIVIDADE: ')
#adiciona a tarefa indicada pelo usuario a lista de tarefas
tarefas.append(atividade)
```

```python
print(tarefas)
```

# INSERINDO VARIOS ITENS NA LISTA ATRAVES DO INPUT

```python
while atividade:
  atividade = input('INSIRA UMA ATIVIDADE: ')
  tarefas.append(atividade)
```

```python
#a estrutura abaixo indica que pegará os itens da posição [0] até o penultimo item da lista [-2]
print(tarefas[:-2])
```

utilização da estrutura FOR para percorrer todos os elementos e fazer o print de cada um deles.

```python
#vamos escrever todos os itens da lista em outro formato

#o for nos permite percorrer todas itens da lista taredas
for tarefa in tarefas:
  print(tarefa)
```

# CRIANDO UMA LISTA COM VALORES NUMERICOS

```python
#ate agora todas nossas listas foram compostas por textos, mas é possivel ter listas com valores  numericos
#no exemplo abaixo temos 3 lojar e seu volume em vendas

#usando numeros
lojas = ['Rio de Janeiro','São Paulo','Curitiba']
vendas = [10000, 20000, 50000]
```

```python
# .sort() -> Ordena os valores do maior para o menor
# ATENÇÃO -> AS DUAS LISTAS NÃO ESTÃO CONECTADAS
lojas.sort()
vendas.sort()
```

```python
#perceba que ao mudarmos a ordem, nao temos ms a rio de janeiro - 10000 / sao paulo - 20000 / curitiba - 50000
print(lojas)
print(vendas)
```

Outra forma de criar uma lista é utilizando o while + input, então até que a informação da caixa de entrada seja vazia o programa irá armazenar as informações em uma lista.

# VAMOS TENTAR UNIR OS ITENS EM UMA TUPLA:
Tupla é um objeto assim como as listas, ao inves de usarmos o [] usaremos () para criar uma dupla

```python
lojas = ['Rio de Janeiro','São Paulo','Curitiba']
vendas = [10000,20000,50000]
```

```python
resultados=[]
#usando range para rodar o for. i é uma variavel auxiliar que irá percorrer a posição 0, 1 e 2 da nossas listas
# range(3) indica que sao 3 interações
for i in range(3):
  tupla=(lojas[i], vendas[i])
  resultados.append(tupla)
```

```python
print(resultados)
```

PARA ACESSARMOS UM ITEM DENTRO DA TUPLA IREMOS 1º INDICAR QUAL A POSIÇÃO DO ITEM DA LISTA E DEPOIS A POSICAO DA TUPLA.
NO EXEMPLO ABAIXO AO FAZERMOS RESULTADOS[1] ESTAMOS ACESSANDO O SEGUNDO ITEM DA LISTA -> ('São Paulo',20000)

AO FAZERMOS resultados[1] [0] ESTAMOS ACESSANDO O PRIMEIRO ITEM DA TUPLA, OU SEJA: São Paulo

```python
print(resultados[1][0])
```
